```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 完整数据定义
data = {
    'Model': ['BERT', 'GPT-2', 'GPT-3', 'Gopher', 'PaLM', 'LLaMA'],
    'Year': [2018, 2019, 2020, 2021, 2022, 2023],
    'Parameters (Billion)': [0.11, 1.5, 175, 280, 540, 65],
    'Dataset Tokens (Billion)': [3.3, 40, 300, 780, 780, 1000],
    'Training FLOPs (Quadrillion)': [0.2, 3.0, 364, 1530, 2040, 150],
    'Zero-shot Accuracy (%)': [80, 85, 88, 90, 92, 89]
}

# 构建 DataFrame
df = pd.DataFrame(data)

# 将 Pandas Series 转为 NumPy 一维数组，避免 FutureWarning
years = df['Year'].to_numpy()
params = df['Parameters (Billion)'].to_numpy()
tokens = df['Dataset Tokens (Billion)'].to_numpy()
flops = df['Training FLOPs (Quadrillion)'].to_numpy()
accuracy = df['Zero-shot Accuracy (%)'].to_numpy()

# 图 1：模型参数量随年份增长（对数刻度）
plt.figure()
plt.plot(years, params, marker='o')
plt.yscale('log')
plt.xlabel('Year')
plt.ylabel('Parameters (Billions, log scale)')
plt.title('Growth of Model Parameter Counts Over Time')
plt.tight_layout()
plt.show()

# 图 2：训练数据集大小随年份增长（对数刻度）
plt.figure()
plt.plot(years, tokens, marker='o')
plt.yscale('log')
plt.xlabel('Year')
plt.ylabel('Dataset Tokens (Billions, log scale)')
plt.title('Growth of Training Dataset Sizes Over Time')
plt.tight_layout()
plt.show()

# 图 3：训练计算量（FLOPs）随年份增长（对数刻度）
plt.figure()
plt.plot(years, flops, marker='o')
plt.yscale('log')
plt.xlabel('Year')
plt.ylabel('Training FLOPs (Quadrillions, log scale)')
plt.title('Estimated Training Compute Requirements Over Time')
plt.tight_layout()
plt.show()

# 图 4：Zero-shot 精度 vs 模型规模（参数量对数刻度）
plt.figure()
plt.scatter(params, accuracy)
plt.xscale('log')
plt.xlabel('Parameters (Billions, log scale)')
plt.ylabel('Zero-shot Accuracy (%)')
plt.title('Zero-shot Accuracy vs Model Size')
plt.tight_layout()
plt.show()
```
